## Plot comparison between different feature selection and classification methods

```python
df2plot = deepcopy(df_ml_af)
df2plot['method'] = [x+'+'+y for x,y in zip(df2plot.feature_selector, df2plot.classification_model)]

PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'black'},
    'medianprops':{'color':'black'},
    'whiskerprops':{'color':'black'},
    'capprops':{'color':'black'}
}

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(15,4), sharey=True)

for k, metric in enumerate(['precision_test','recall_test','f1_test','accuracy_test']):
    order = list(df2plot.groupby('method')[metric].agg(np.median).sort_values().index)
    _ = sns.boxplot(x="method", y=metric, data=df2plot, showfliers=False, **PROPS, order=order, ax=ax[k])
    _ = sns.stripplot(x="method", y=metric, data=df2plot, order=order, ax=ax[k], color='gray')
    _ = ax[k].set_xticks(ax[k].get_xticks())
    _ = ax[k].set_xticklabels(ax[k].get_xticklabels(), rotation=45)
    _ = ax[k].set_xlabel('')
    _ = ax[k].set_ylabel('Performance')
    _ = ax[k].set_title(metric.replace("_test",""), fontsize=12)
    _ = ax[k].set_ylim([-0.1,1.1])

plt.tight_layout()
plt.savefig("output/comparison_between_feature_selectors_and_classifiers.png", dpi=600, bbox_inches='tight')
```

## Plot comparison of F1 score between XGBoost model with all features and SNV-only model

```python
# machine learning model using all features
df2plot_1 = deepcopy(df_ml_af)
df2plot_1 = df2plot_1[(df2plot_1.feature_selector=='Logit') & (df2plot_1.classification_model=='XGB')]
df2plot_1['method'] = 'all features'

# machine learning model using SNV only
df2plot_2 = deepcopy(df_ml_snv)
df2plot_2['method'] = 'SNV only'

# merge evaluation model results
df2plot = pd.concat([df2plot_1, df2plot_2]).reset_index(drop=True)

PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'black'},
    'medianprops':{'color':'black'},
    'whiskerprops':{'color':'black'},
    'capprops':{'color':'black'}
}

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(2,4))
_ = sns.boxplot(x="method", y='f1_test', data=df2plot, showfliers=False, **PROPS, order=['all features','SNV only'], ax=ax)
_ = sns.stripplot(x="method", y='f1_test', data=df2plot, order=['all features','SNV only'], ax=ax, color='gray')
_ = ax.set_xticks(ax.get_xticks())
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
_ = ax.set_xlabel('')
_ = ax.set_ylabel('F1 score', fontsize=12)
_ = ax.set_ylim([-0.1,1.1])
    
_ = plt.savefig("output/comparison_of_XBG_between_all_features_and_snv_only.png", dpi=600, bbox_inches='tight')
```

## Plot comparison of F1 score between machine learning model and alternative models (machine learning model with shuffling, ANI-based model, phylogeny-based model)

```python
# machine learning model using all features
df2plot_1 = deepcopy(df_ml_af)
df2plot_1 = df2plot_1[(df2plot_1.feature_selector=='Logit') & (df2plot_1.classification_model=='XGB')]
df2plot_1['method'] = 'Logit+XGB'

# machine learning model using shuffled heteroresistance labels
df2plot_2 = deepcopy(df_ml_af_shuffled)
df2plot_2['method'] = ['Logit+XGB, shuffled, %d'%(repeat) for repeat in df2plot_2.repeat]

# ANI-based model
df2plot_3 = deepcopy(df_ani)
df2plot_3['method'] = 'ANI'

# phylogeny-based model
df2plot_4 = deepcopy(df_phylo)
df2plot_4['method'] = 'Phylogeny'

# merge evaluation model results
columns2keep = ['method', 'accuracy_test', 'precision_test', 'recall_test', 'f1_test']
df2plot = pd.concat([df2plot_1[columns2keep], df2plot_2[columns2keep], df2plot_3[columns2keep], df2plot_4[columns2keep]]).reset_index(drop=True)

PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'black'},
    'medianprops':{'color':'black'},
    'whiskerprops':{'color':'black'},
    'capprops':{'color':'black'}
}

fig, ax = plt.subplots(figsize=(4,4))
order = ['Logit+XGB', 'Logit+XGB, shuffled, 1', 'Logit+XGB, shuffled, 2', 'Logit+XGB, shuffled, 3', 'ANI', 'Phylogeny']
_ = sns.boxplot(x="method", y='f1_test', data=df2plot, showfliers=False, **PROPS, order=order, ax=ax)
_ = sns.stripplot(x="method", y='f1_test', data=df2plot, order=order, ax=ax, color='grey')
_ = ax.set_ylim([-0.1,1.1])
_ = ax.set_xticks(ax.get_xticks())
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
_ = ax.set_xlabel('')
_ = ax.set_ylabel('F1 score')

plt.tight_layout()
_ = plt.savefig("output/comparison_between_machine_learning_and_alternative_models.png", dpi=600, bbox_inches='tight')
```

## Feature prevalence analysis (Logit+XBGoost)

```python
df_feature_rank = pd.read_csv("data/logit_feature_rank.csv")
df_ml_af_logit_xgb = df_ml_af[(df_ml_af.feature_selector=='Logit') & (df_ml_af.classification_model=='XGB')]

res = []
for idx, features in zip(df_ml_af_logit_xgb.iteration, df_ml_af_logit_xgb.selected_features):
    feature_list = features.split(',')
    for feature in feature_list:
        res.append([idx, feature.split('(')[0].strip(), float(feature.split('(')[1].split(')')[0].strip())])
df_selected_features = pd.DataFrame(res, columns=['Iteration', 'Feature', 'Score'])
df_selected_features_wide = pd.pivot_table(df_selected_features, index='Feature', columns='Iteration', values='Score').fillna(0)
df_selected_features = df_selected_features_wide.stack().reset_index().rename({0:'Score'}, axis=1)
df_selected_features = pd.merge(df_selected_features, df_feature_rank, left_on='Feature', right_on='Feature', how='left')
df_selected_features['NewFeature'] = ['%s (%d)'%(f.replace('Pattern','Feature'), r) for f,r in zip(df_selected_features.Feature, df_selected_features.Rank)]

fig, ax = plt.subplots(figsize=(10,4))
grouped = df_selected_features.groupby(['NewFeature'])['Score'].mean().sort_values(ascending=False)
_ = sns.boxplot(x="NewFeature", y="Score", data=df_selected_features, order=grouped.index, ax=ax, fliersize=4, color='salmon', saturation=0.75)
_ = ax.set_xticks(ax.get_xticks())
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
_ = ax.set_xlabel('')
_ = ax.set_ylabel('Importance score')

_ = plt.savefig("output/feature_prevalence_analysis_logit_xgboost.png", dpi=600, bbox_inches='tight')
```

## Frequency of false positive and false negatives

```python
df_splits = pd.read_csv("data/train_test_splits.csv")
df_splits.head()
```

```python
false_positive = {}
for fp_isolates in df_ml_af_logit_xgb.false_positive_test:
    if str(fp_isolates) != 'nan':
        for isolate in fp_isolates.split(','):
            if isolate in false_positive.keys():
                false_positive[isolate] += 1
            else:
                false_positive[isolate] = 1
df_fp = pd.DataFrame.from_dict(false_positive, orient='index').reset_index().rename({'index':'Isolate',0:'Count'},axis=1)

tested = []
for isolate in df_fp.Isolate:
    tested.append(len(df_splits[df_splits.test_set_isolates.str.contains(isolate)]))
df_fp['Ntest'] = tested
df_fp['Percentage'] = df_fp.Count/df_fp.Ntest

df_fp['Isolate2'] = ['%s (%d)'%(x,y) for x,y in zip(df_fp.Isolate, df_fp.Ntest)]
df_fp.head()
```

```python
false_negative = {}
for fn_isolates in df_ml_af_logit_xgb.false_negative_test:
    if str(fn_isolates) != 'nan':
        for isolate in fn_isolates.split(','):
            if isolate in false_negative.keys():
                false_negative[isolate] += 1
            else:
                false_negative[isolate] = 1
df_fn = pd.DataFrame.from_dict(false_negative, orient='index').reset_index().rename({'index':'Isolate',0:'Count'},axis=1)

tested = []
for isolate in df_fn.Isolate:
    tested.append(len(df_splits[df_splits.test_set_isolates.str.contains(isolate)]))
df_fn['Ntest'] = tested
df_fn['Percentage'] = df_fn.Count/df_fn.Ntest

df_fn['Isolate2'] = ['%s (%d)'%(x,y) for x,y in zip(df_fn.Isolate, df_fn.Ntest)]
df_fn.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4), sharey=True)

# false positive
_ = sns.barplot(x='Isolate2', y='Percentage', data=df_fp, order=df_fp.sort_values('Percentage', ascending=False).Isolate2, ax=ax[0], color='orange', linewidth=1, edgecolor='k')
_ = ax[0].set_xticks(ax[0].get_xticks())
_ = ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=90)
_ = ax[0].set_title('False positives')
_ = ax[0].set_ylim([0,1.1])
_ = ax[0].set_yticks([0,0.2,0.4,0.6,0.8,1])
_ = ax[0].set_yticklabels([0,20,40,60,80,100])
_ = ax[0].set_xlabel('')
_ = ax[0].set_ylabel('% misclassifications')

_ = sns.barplot(x='Isolate2', y='Percentage', data=df_fn, order=df_fn.sort_values('Percentage', ascending=False).Isolate2, ax=ax[1], color='skyblue', linewidth=1, edgecolor='k')
_ = ax[1].set_xticks(ax[1].get_xticks())
_ = ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=90)
_ = ax[1].set_title('False negatives')
_ = ax[1].set_ylim([0,1.1])
_ = ax[1].set_yticks([0,0.2,0.4,0.6,0.8,1])
_ = ax[1].set_yticklabels([0,20,40,60,80,100])
_ = ax[1].set_xlabel('')
_ = ax[1].set_ylabel('% misclassifications')

_ = plt.tight_layout()
_ = plt.savefig("output/FP_FN_analysis.png", dpi=600, bbox_inches='tight')
```

## Impact of undersampling and oversampling

```python
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(12,4), sharey=True)

df_ml_af2 = deepcopy(df_ml_af)
df_ml_af2 = df_ml_af2[(df_ml_af2.feature_selector=='Logit') & (df_ml_af2.classification_model=='XGB')]
df_ml_af2['strategy'] = 'no resampling'
df2plot = pd.concat([df_ml_af2[list(df_ml_af_resampling.columns)], df_ml_af_resampling], axis=0)

PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'black'},
    'medianprops':{'color':'black'},
    'whiskerprops':{'color':'black'},
    'capprops':{'color':'black'}
}

for k, metric in enumerate(['precision','recall','f1','accuracy']):
    _ = sns.boxplot(x="strategy", y=metric+"_test", data=df2plot, ax=ax[k], showfliers=False, **PROPS)
    _ = sns.stripplot(x="strategy", y=metric+"_test", data=df2plot, ax=ax[k], color='gray')
    _ = ax[k].set_xlabel('')
    _ = ax[k].set_ylabel('Performance')
    _ = ax[k].set_xticks(ax[k].get_xticks())
    _ = ax[k].set_xticklabels(ax[k].get_xticklabels(), rotation=45, ha='right')
    _ = ax[k].set_ylim([0.1,1.1])
    _ = ax[k].set_title(metric)
    
_ = plt.tight_layout()
_ = plt.savefig("output/comparison_between_resampling_techniques.png", dpi=600, bbox_inches='tight')
```
