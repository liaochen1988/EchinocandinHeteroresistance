```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from statsmodels.stats.multitest import multipletests
```

```python
df = pd.read_csv("Moran_test_results_10182023.csv", index_col=0).T
df.loc['HR']
```

# Features

```python
df['Padj'] = multipletests(df.Pvalue, method='bonferroni')[1]
df['tmp'] = -df['Observed']
df = df.sort_values(['Padj','tmp']).drop('tmp', axis=1)
df.index.name = 'Isolate'
df.to_csv("Moran_test_results_10182023_processed.csv")
df.head()
```

```python
df[df.Padj!=0].min()
```

```python
df.Observed.min(), df.Observed.max()
```

# number of HR correlated features that are also correlated with phylogeny

```python
df_logit = pd.read_csv("../logistic_regression/logistic_univariate.csv", index_col=0)
df_logit = pd.merge(df_logit, df, left_index=True, right_index=True, how='inner').rename({'Padj_x':'Padj_logit','Padj_y':'Padj_moranI'}, axis=1)

print('# HR correlated features = ', len(df_logit[df_logit.Padj_logit < 0.05]))
print('# HR and phylo correlated features = ', len(df_logit[(df_logit.Padj_logit < 0.05) & (df_logit.Padj_moranI<0.05)]))
```

# Plot relationships of features that are correlated with phylogeny and that are correlated with phenotype

```python
df2plot = pd.merge(
    pd.read_csv("../logistic_regression/logistic_univariate.csv", index_col=0)[['Coef','Padj']].rename({'Coef':'Coef_logit', 'Padj':'P_logit'}, axis=1),
    df[['Observed','Expected','Padj']].rename({'Padj':'P_moran'}, axis=1),
    left_index=True,
    right_index=True,
    how='inner')
df2plot['Coef_MoranI_norm'] = df2plot['Observed']-df2plot['Expected']
df2plot['Abs_Coef_logit'] = np.abs(df2plot['Coef_logit'])
df2plot = df2plot[df2plot.Abs_Coef_logit<5]
df2plot['Sign of logic coef'] = 'Negative'
df2plot.loc[df2plot.Coef_logit>0 ,'Sign of logic coef'] = 'Positive'

r,p = stats.pearsonr(df2plot.Coef_MoranI_norm, df2plot.Abs_Coef_logit)

fig, ax = plt.subplots(figsize=(5,5))
_ = sns.scatterplot(x="Coef_MoranI_norm", y="Abs_Coef_logit", data=df2plot, ax=ax, hue='Sign of logic coef', s=16, edgecolor=None, palette='Set2')
_ = sns.regplot(x="Coef_MoranI_norm", y="Abs_Coef_logit", data=df2plot, ax=ax, scatter=False, color='k')
_ = ax.set_xlabel("Phylogenetic correlation", fontsize=12)
_ = ax.set_ylabel("Heteroresistance correlation", fontsize=12)
_ = ax.set_title("r = %2.2f, P = %2.1e"%(r,p))

plt.tight_layout()
plt.rcParams['svg.fonttype'] = 'none'
plt.savefig("scatterplot_phylocorr_hrcorr.png", dpi=600, bbox_inches='tight')
```
