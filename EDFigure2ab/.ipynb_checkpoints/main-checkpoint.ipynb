## plot consistency in PAP survival curves across replicates

```python
df_repeats = pd.read_csv("data/PAP_repeats_15_blood_isolates.csv", index_col=0)
df_repeats = df_repeats[[col for col in df_repeats.columns if '_4' not in col]]

fig, ax = plt.subplots(nrows=3, ncols=5, figsize=(8,4), sharex=True, sharey=True)

all_isolates = list(set([col.split('_')[0] for col in df_repeats.columns]))
assert len(all_isolates)==15

for k,isolate in enumerate(all_isolates):
    krow = int(k/5)
    kcol = k % 5
    for m in ['_1','_2','_3']:
        _ = ax[krow, kcol].plot(np.arange(7), np.log10(df_repeats[isolate+m]), marker='s', markersize=4)
    _ = ax[krow, kcol].set_ylim([-4.5,2.5])
    _ = ax[krow, kcol].set_yticks([-4,-3,-2,-1,0,1,2])
    _ = ax[krow, kcol].set_yticklabels([0.0001,0.001,0.01,0.1,1,10,100])
    _ = ax[krow, kcol].set_title('MSK'+isolate)
    _ = ax[krow, kcol].set_xlim([-0.5,6.5])
    _ = ax[krow, kcol].set_xticks([0,1,2,3,4,5,6])
    _ = ax[krow, kcol].set_xticklabels([0,1,2,4,8,16,32])
    
plt.tight_layout()
fig.savefig("output/PAP_repeats_survival_curve.png", dpi=600, bbox_inches='tight')
```

## Plot consistency in heteroresistance classification across replicates

```python
df_repeats = pd.read_csv("data/PAP_repeats_15_blood_isolates.csv", index_col=0)
df_repeats_T = df_repeats.T
hr_124  = list(df_repeats_T[((df_repeats_T[1]<50)&(df_repeats_T[1]>1e-4)) & ((df_repeats_T[2]<50)&(df_repeats_T[2]>1e-4)) & ((df_repeats_T[4]<50)&(df_repeats_T[4]>1e-4))].index)
hr_248  = list(df_repeats_T[((df_repeats_T[2]<50)&(df_repeats_T[2]>1e-4)) & ((df_repeats_T[4]<50)&(df_repeats_T[4]>1e-4)) & ((df_repeats_T[8]<50)&(df_repeats_T[8]>1e-4))].index)
hr_4816 = list(df_repeats_T[((df_repeats_T[4]<50)&(df_repeats_T[4]>1e-4)) & ((df_repeats_T[8]<50)&(df_repeats_T[8]>1e-4)) & ((df_repeats_T[16]<50)&(df_repeats_T[16]>1e-4))].index)
df_repeats_T['HR_124'] = False
df_repeats_T.loc[hr_124, 'HR_124'] = True
df_repeats_T['HR_248'] = False
df_repeats_T.loc[hr_248, 'HR_248'] = True
df_repeats_T['HR_4816'] = False
df_repeats_T.loc[hr_4816, 'HR_4816'] = True
df_repeats_T['HR'] = False
df_repeats_T.loc[df_repeats_T[df_repeats_T.HR_124 | df_repeats_T.HR_248 | df_repeats_T.HR_4816].index,'HR'] = True
df_heatmap = df_repeats_T[['HR']].reset_index().rename({'index':'isolate_Repeat_no'}, axis=1)
df_heatmap['isolate_no'] = ['MSK'+x.split('_')[0] for x in df_heatmap.isolate_Repeat_no]
df_heatmap['Repeat'] = [int(x.split('_')[1]) for x in df_heatmap.isolate_Repeat_no]
df_heatmap = pd.pivot_table(df_heatmap, index='isolate_no', columns='Repeat', values='HR')

# keep 3 replicates (not all isolates have four replicates)
df_heatmap = df_heatmap[[1,2,3]]

fig, ax = plt.subplots(figsize=(2,8))
colors = ["lightblue", "tomato"] 
cmap = LinearSegmentedColormap.from_list('Custom', colors, len(colors))
_ = sns.heatmap(df_heatmap, vmin=0, vmax=1, ax=ax, mask=df_heatmap.isnull(), cmap=cmap, cbar_kws={"shrink": .5}, linewidth=1, linecolor='k')
_ = ax.set_ylabel('') 

colorbar = ax.collections[0].colorbar
colorbar.set_ticks([0.25,0.75])
colorbar.set_ticklabels(['S', 'HR'])

fig.savefig("output/PAP_repeats_classification_consistency.png", dpi=600, bbox_inches='tight')
```
