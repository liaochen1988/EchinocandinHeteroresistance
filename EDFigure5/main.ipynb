```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from scipy.signal import find_peaks
```

```python
from platform import python_version
print(python_version())
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15,6))
isolates = ['E66',        'MSK2384',      'CDC340',       'UWM1206',     'GL62']
orfs =     ['CPAR2_106940','CPAR2_104600','CPAR2_104610','CPAR2_501360','CPAR2_600430']

df_reads_per_copy = pd.read_csv("data/number_reads_per_chromosome_copy.csv", index_col=0)
dict_reads_per_copy = {x:y for x,y in zip(df_reads_per_copy.index, df_reads_per_copy.ReadsPerCopy)}

for k in np.arange(len(isolates)):
    k1 = int(k/3)
    k2 = k % 3
    isolate = isolates[k]
    orf = orfs[k]
    
    df_hist2 = pd.read_csv("data/%s.read_density.hist.txt.gz"%(isolate), compression="gzip", sep="\t", header=None)
    df_hist2 = df_hist2[df_hist2[3] == orf]
    df_hist2 = df_hist2[[3,4,5,6]]
    df_hist2.columns = ['Orf','Depth','Count','Length']
    df_hist2[['Depth','Count','Length']] = df_hist2[['Depth','Count','Length']].astype(int)
    df_hist2['Isolate'] = isolate

    min_depth = np.min(df_hist2.Depth)
    max_depth = np.max(df_hist2.Depth)
    orf_length = df_hist2.Length.mean()
    df_tmp = pd.concat([pd.DataFrame([[orf,min_depth-1,1,orf_length, isolate]], columns=df_hist2.columns), df_hist2])

    inputd = []
    for x,y in zip(df_tmp.Depth, df_tmp.Count):
        inputd.extend([x]*int(y))
    kernel = gaussian_kde(inputd)
    xs = np.arange(min_depth-1, max_depth+1)     

    factor=1
    peaks, _ = find_peaks(kernel(xs), distance=1e10)
    if len(peaks)>0:
        dominant_cnv = xs[peaks[0]]/dict_reads_per_copy[isolate]
        if np.round(dominant_cnv) != 0:
            factor = dominant_cnv/np.round(dominant_cnv)
    adjusted_reads_per_copy = dict_reads_per_copy[isolate]*factor

    # plot histogram
    _ = ax[k1,k2].plot(df_tmp.iloc[1:].Depth/adjusted_reads_per_copy, df_tmp.iloc[1:].Count, color='silver')
    _ = ax[k1,k2].set_title('%s, %s'%(isolate, orf))
    _ = ax[k1,k2].set_yscale('log')
    _ = ax[k1,k2].set_ylabel('Number of base pairs')
    
    # plot kernel
    ax2 = ax[k1,k2].twinx()
    _ = ax2.plot(xs[1:]/adjusted_reads_per_copy, kernel(xs[1:]), color='red')

    # find peaks
    peaks, _ = find_peaks(kernel(xs), distance=adjusted_reads_per_copy*0.9)
    peaks.sort()
    reads_peaks = xs[peaks]
    for m,reads_peak in enumerate(reads_peaks):
        if m==0:
            left_boundary = 0
        else:
            left_boundary = (reads_peaks[m]+reads_peaks[m-1])/2
        if m==len(reads_peaks)-1:
            right_boundary = 1e10
        else:
            right_boundary = (reads_peaks[m]+reads_peaks[m+1])/2
        peak_contribution = df_tmp[(df_tmp.Depth>=left_boundary) & (df_tmp.Depth<=right_boundary)].Count.sum()/orf_length
        if peak_contribution > 1:
            peak_contribution = 1

        if peak_contribution >= 0.2:
            cnv_peak = reads_peak/adjusted_reads_per_copy
            height = np.interp(cnv_peak, xs/adjusted_reads_per_copy, kernel(xs))
            _ = ax2.plot([cnv_peak, cnv_peak],[0, height], '--', color='red', label='cn=%d (%2.2f%%)'%(np.round(cnv_peak),peak_contribution*100))
    _ = ax2.legend()
    _ = ax2.set_ylabel('Kernel density')

_ = ax[1,2].set_visible(False)

_ = plt.tight_layout()
_ = plt.savefig('output/cnv_algo_illustration.png', dpi=600, bbox_inches="tight") 
```
