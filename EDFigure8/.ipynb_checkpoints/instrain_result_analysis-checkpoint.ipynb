```python
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.patches import Patch
```

## Load collection site of isolates

```python
df_meta = pd.read_csv("data/isolates.csv")[['Isolate','Institute','Location']].set_index('Isolate')
df_meta.head()
```

## Load popANI values from inStrain output

```python
df_instrain = pd.read_csv('data/instrainCompare_comparisonsTable_0507.tsv', sep='\t')
df_instrain_1 = df_instrain[['name1','name2','popANI']]
df_instrain_2 = df_instrain[['name2','name1','popANI']]
df_instrain_2.columns = ['name1','name2','popANI']
df_augmented = pd.concat([df_instrain_1, df_instrain_2], axis=0)
df_augmented = df_augmented.groupby(['name1','name2'])['popANI'].agg(np.mean).reset_index()                           
df_instrain_popANI = pd.pivot_table(df_augmented, index="name1", columns="name2", values="popANI").fillna(1)
df_instrain_popANI = df_instrain_popANI.loc[df_meta.index, df_meta.index]
df_instrain_popANI.to_csv("output/popANI_wide_format.csv")
df_instrain_popANI.head()
```

## Hierachical clustering

```python
df_instrain_popANI_binary = (df_instrain_popANI>0.99999).astype(int)
d = sch.distance.pdist(df_instrain_popANI_binary, metric='jaccard')
L = sch.linkage(d, method='average')
lut = dict(zip(['USA','China','Europe'], [(255/255,0,0),(70/255,130/255,180/255),(50/255,205/255,50/255)]))
location = list(df_meta.Location)
row_colors = [lut[loc] for loc in location]
g = sns.clustermap(df_instrain_popANI_binary, 
                   cmap=colors.ListedColormap(['white', 'slategray']), 
                   linewidth=0.01, 
                   linecolor=None,
                   figsize=(10,10),
                   row_linkage=L,
                   col_linkage=L,
                   yticklabels=True,
                   xticklabels=True,
                   row_colors=row_colors,
                   col_colors=row_colors
                  )
g.ax_heatmap.set_yticklabels(g.ax_heatmap.get_ymajorticklabels(), fontsize = 6)
g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xmajorticklabels(), fontsize = 6)
g.cax.set_visible(False)
plt.savefig('output/hierachical_clustering_popANI_%.10g.png'%(0.99999), dpi=600)
```

## Identify distinct clusters of identifical isolates at different popANI cutoffs

```python
# the value of a threshold is encoded by the number of digit "9" it contains
# e.g., 2.0->99%, 3.0->99.9%, 4.0->99.99%, ..., 10.0->99.99999999%
thresholds = np.arange(2,10.5,0.5)
number_of_clusters = []

for thres in thresholds:
    df_popANI_thres = pd.read_csv("output/popANI_wide_format.csv", index_col=0)
    df_popANI_thres = (df_popANI_thres>(1-10**(-thres))).astype(int)
    df_clusters_final = pd.DataFrame(index=df_popANI_thres.index, columns=['ClusterName'])
    cluster_name = 0
    
    resume = True
    while resume:
        if len(df_popANI_thres) == 0:
            break
        
        # find the biggest cluster where every two strains are the same in pairwise comparison
        d = sch.distance.pdist(df_popANI_thres, metric='jaccard')
        L = sch.linkage(d, method='average')
        found_a_cluster=False
        
        for p in np.linspace(0.99,0,100):
            clusters = sch.fcluster(L, p*d.max(), 'distance')
            df_clusters = pd.DataFrame.from_dict(dict(zip(df_popANI_thres.index, clusters)), orient='index').rename({0:'cluster'}, axis=1)
            unique_clusters = list(set(df_clusters.cluster))
            number_of_strains_in_clusters = [len(list(df_clusters[df_clusters.cluster==cluster].index)) for cluster in unique_clusters] # rank these clusters
            unique_clusters_sorted = [x for _, x in sorted(zip(number_of_strains_in_clusters, unique_clusters), key=lambda pair: pair[0], reverse=True)]
    
            for cluster in unique_clusters_sorted:
                strains_of_this_cluster = list(df_clusters[df_clusters.cluster==cluster].index)
                strains_not_in_this_cluster = list(set(df_clusters.index) - set(strains_of_this_cluster))
                # if it comprises a single strain, skip
                if len(strains_of_this_cluster) > 1:
                    # only for strains never found before
                    df_popANI_thres_subset = df_popANI_thres.loc[strains_of_this_cluster, strains_of_this_cluster]
                    if np.sum(df_popANI_thres_subset.values) == len(df_popANI_thres_subset)**2:
                        # found a cluster 
                        df_clusters_final.loc[strains_of_this_cluster, 'ClusterName'] = cluster_name
                        cluster_name += 1
                        df_popANI_thres = df_popANI_thres.loc[strains_not_in_this_cluster, strains_not_in_this_cluster]
                        
                        found_a_cluster=True
                        break
            
            if found_a_cluster:
                break
        
        if found_a_cluster:
            resume = True
        else:
            resume = False
    
    # each of the remaining strains represents an individual cluster
    for s in df_clusters_final[df_clusters_final.ClusterName.isnull()].index:
        df_clusters_final.loc[s,'ClusterName'] = cluster_name
        cluster_name += 1
    
    # save to file
    df_clusters_final.index.name = 'Isolate'
    df_clusters_final = df_clusters_final.sort_values('ClusterName')
    df_s2p_mapping=pd.read_csv("data/sample_to_patient_mapping.csv", index_col=0)
    df_clusters_final = pd.merge(df_clusters_final, df_s2p_mapping, left_index=True, right_index=True, how='left') # add patient ID
    df_clusters_final.to_csv('output/strain_clusters_popANI_cutoff_%2.1f.csv'%(thres))

    # record number of clusters found at this cutoff
    number_of_clusters.append(len(set(df_clusters_final.ClusterName)))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
_ = ax.plot(thresholds, number_of_clusters, 'ko-', markerfacecolor='w')
_ = ax.set_xticks([2,3,4,5,6,7,8,9,10])
_ = ax.set_xticklabels(['99%','99.9%','99.99%','99.999%','99.9999%','99.99999%','99.999999%','99.9999999%','99.99999999%'], rotation=45, ha='right')
_ = ax.set_xlabel('popANI threshold')
_ = ax.set_ylabel('# distinct clusters')
plt.savefig('output/number_of_clusters_at_diff_popANI_thres.png', dpi=600)
```
