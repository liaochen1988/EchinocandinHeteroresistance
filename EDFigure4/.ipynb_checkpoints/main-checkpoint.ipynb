```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from platform import python_version
print(python_version())
```

## Load average copy number per 5kb window

```python
df_5kb_bin = pd.read_csv('data/5kb_bin_copy_number.csv')
df_5kb_bin.head()
```

## Load chromosome start and end position

```python
df_chr = pd.read_csv("data/chromosome_start_end_pos.csv", index_col=0)
df_chr.head()
```

## Load isolates with aneuploidy

```python
df_aneu = pd.read_csv("data/aneuploidy_cnv.csv", index_col=0)
df_aneu.head()
```

## Plot copy number fluctuations across chromosomes

```python
fig, ax = plt.subplots(nrows=3, ncols=5, figsize=(15,5), sharex=True, sharey=True)

curr_index = 0
max_copy_number = 6
scaling = 1e6
for curr_index, s in enumerate(df_aneu.index):
    df2 = df_5kb_bin[df_5kb_bin.Isolate==s]
    offset = 0
    row = int(curr_index/5)
    col = curr_index%5
    for c in list(df_chr.index):
        df3 = df2[df2.Chromosome==c].sort_values('BinStartPos')
        _ = ax[row,col].plot((df3.BinStartPos+offset)/scaling, df3.CopyNumber, color='deepskyblue', label=s)
        if c=='cpar_Chr_1':
            _ = ax[row,col].legend(fontsize=14, loc='upper left')
        _ = ax[row,col].fill_between((df3.BinStartPos+offset)/scaling, df3.CopyNumber, color='skyblue', alpha=0.5)
        offset = df_chr.loc[c,'END']+offset
        if c != 'cpar_Chr_8':
            _ = ax[row,col].plot([offset/scaling,offset/scaling],[0,max_copy_number],'--',color='gray',linewidth=1)

    _ = ax[row,col].plot([0,offset/scaling],[1,1],'--',color='gray',linewidth=1)
    _ = ax[row,col].plot([0,offset/scaling],[3,3],'--',color='gray',linewidth=1)
    if col==0:
        _ = ax[row,col].set_ylabel('Copy number', fontsize=12)
    if row==2:
        _ = ax[row,col].set_xlabel('Genome (Mbp)', fontsize=12)
    _ = ax[row,col].set_xlim([0,offset/scaling])
    _ = ax[row,col].set_ylim([0,max_copy_number])
    _ = ax[row,col].set_yticks(np.arange(max_copy_number+1))
    _ = ax[row,col].tick_params(axis='both', labelsize=12)

    curr_index += 1
    _ = ax[row,col].legend_.remove()
    _ = ax[row,col].set_title('%s, copy=%2.2f'%(s, df_aneu.loc[s,'CopyNumberOfAbnormalRegion']))
    _ = ax[row,col].plot([0,offset/scaling], [2,2], 'k--')
        
plt.tight_layout()
_ = plt.savefig("output/genome_wide_chromosome_copy_number.png", dpi=600, bbox_inches="tight")
```
